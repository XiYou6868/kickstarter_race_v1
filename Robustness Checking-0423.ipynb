```python
# Select rows with inconsistent project IDs
inconsistent_projects_df = df_creator[df_creator['ProjectID'].isin(duplicate_project_ids)]

# Display the selected rows
inconsistent_projects_df.info()
inconsistent_projects_df['ProjectID'].nunique()
```

```python
inconsistent_projects_df.head()
```

```python
# Columns to check for consistency
columns_to_check = ['CreatorName', 'EndDate','Backer','RewardNum','Video_in_Description']

# Initialize a dictionary to store inconsistent content for each ProjectID
inconsistent_content = {}

# Step 1: Group the DataFrame by ProjectID
grouped_creator = inconsistent_projects_df.groupby('ProjectID')

# Step 2: Iterate through each group to check for inconsistent values
for project_id, group in grouped_creator:
    # Check if there are inconsistent values within the group for the specified columns
    if not group[columns_to_check].apply(lambda x: x.nunique()).eq(1).all():
        # Store the inconsistent values for the specified columns
        inconsistent_content[project_id] = group[columns_to_check]

# Display inconsistent content for each ProjectID
for project_id, content in inconsistent_content.items():
    print("ProjectID:", project_id)
    print(content)
    print()
```
